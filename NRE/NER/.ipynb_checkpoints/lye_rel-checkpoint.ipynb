```python
import sys
import torch

sys.path.append('../')
from rel_extract.extractor import RelationExtractor
from rel_extract.model import RelTaggerModel
from transformers import BertModel
from transformers import BertTokenizer
import os
import pandas as pd
import json
```

```python
_filename = 'save_bert_large_with_squad0'
UNCASED = './bert-large-uncased-whole-word-masking-finetuned-squad'
VOCAB = 'vocab.txt'
pretrained_weights = 'bert-large-uncased-whole-word-masking-finetuned-squad'
model_class = BertModel
tokenizer_class = BertTokenizer
```

```python
tokenizer = tokenizer_class.from_pretrained(pretrained_weights)
```

```python
UNCASED = 'lyh_bert_model'
language_model = model_class.from_pretrained(UNCASED)
```

```python
def sentence_list_re(s_l, model):
    """
    input: list of result in dict
    output: entities and relation in 
    """
    triples = []
    for i in range(len(s_l)):
        re = model.rank(text = s_l[i]['sentence'], head = s_l[i]['head'], tail = s_l[i]['tail'])[0]
        triple = [s_l[i]['head'], re[0], s_l[i]['tail']]
        triples.append(triple)
    return triples
```
