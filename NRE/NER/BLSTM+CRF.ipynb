```python
from model.data_utils import CoNLLDataset
from model.ner_model import NERModel
from model.config import Config
import pandas as pd
```

```python
config = Config()
model = NERModel(config)
model.build()
model.restore_session(config.dir_model)
```

```python
def S_NER(input_sentence, input_model):
	sentence = input_sentence
	model = input_model
	words_raw = sentence.strip().split(" ")
	preds = model.predict(words_raw)
	return preds
```

```python
s = "I love China"
```

```python
S_NER(s, model)
```

```python
with open("data/CoNLL03test.txt", 'r') as f:
    sentences_test = []
    sentence = []
    entity_test = []
    for line in f:
        if len(line) == 0 or line.startswith('-DOCSTART') or line[0] == "\n":
            if len(sentence)>0:
                sentences_test.append(sentence)
                sentence = []
            continue
        splits = line.split(' ')
        sentence.append((splits[0],splits[1] ,splits[-1].strip('\n')))
        entity_test.append((splits[0],splits[1] ,splits[-1].strip('\n')))
        #df_train = df_train.append([{'token': splits[0], 'postag':splits[1], 'entity':splits[-1]}])
    if (len(sentence)> 0):
        sentences_test.append(sentence)
        sentence = []
df_test = pd.DataFrame(entity_test, columns = ["token", "postag", "entity"])
```

```python
p_sentence_test = []

for i in range(len(sentences_test)):
    w = []
    for j in range(len(sentences_test[i])):
        w.append(sentences_test[i][j][0])
    s = " ".join(w)
    p_sentence_test.append(s)
```

```python
result_list = []
for i in range(len(p_sentence_test)):
    result = S_NER(p_sentence_test[i], model)
    result_list.append(result)
```

```python
r = []
for i in range(len(result_list)):
    r += result_list[i]
```
